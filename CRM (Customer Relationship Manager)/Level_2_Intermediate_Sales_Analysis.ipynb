```python
from sqlalchemy import create_engine
import pandas as pd
```

```python
engine = create_engine('mysql+pymysql://root:password@127.0.0.1/crm_sales')
```

### Q1. Display all products with their series and sales price.

```python
query = f"""select * from products;"""
table_data = pd.read_sql(query, engine)
display(table_data)
```

### Q2. Find the average sales price of products in each series.

```python
query = f"""select series,avg(sales_price) from products group by series;"""
table_data = pd.read_sql(query, engine)
display(table_data)
```

### Q3. Join sales_pipeline with products and show each opportunity‚Äôs product name, deal stage, and sales price.

```python
query = f"""select sp.opportunity_id,p.product,sp.deal_stage,p.sales_price from sales_pipeline as sp join products as p on sp.product = p.product;"""
table_data = pd.read_sql(query, engine)
display(table_data)
```

### Q4. Show all deals handled by Darcel Schlecht along with their status (Won/Lost) and close value.

```python
query = f"""select sales_agent,deal_stage,close_value from sales_pipeline where sales_agent = 'Darcel Schlecht';"""
table_data = pd.read_sql(query, engine)
display(table_data)
```

### Q5. Find the total number of deals per deal_stage (Won, Lost, Engaging).

```python
query = f"""select deal_stage,count(deal_stage) from sales_pipeline group by deal_stage;"""
table_data = pd.read_sql(query, engine)
display(table_data)
```

### Q6. List all Won deals with close_value greater than 1000.

```python
query = f"""select * from sales_pipeline where deal_stage = 'Won' and close_value > 1000 ;"""
table_data = pd.read_sql(query, engine)
display(table_data)
```

### Q7. Find top 5 products that generated the highest total deal value.

```python
query = f"""select product,deal_stage,close_value from sales_pipeline order by close_value desc limit 5 ;"""
table_data = pd.read_sql(query, engine)
display(table_data)
```

### Q8. Show the total revenue from all ‚ÄúWon‚Äù deals.    

```python
query = f"""select sum(ac.revenue) as Total_revenue,sp.deal_stage from accounts as ac
            join sales_pipeline as sp on ac.account = sp.account
            where sp.deal_stage = 'Won';"""
table_data = pd.read_sql(query, engine)
display(table_data)
```

### Q9. Find the total number of Won deals per sales agent.

```python
query = f"""select sales_agent,count(deal_stage) as Total_Won,deal_stage from sales_pipeline where deal_stage = 'Won' group by sales_agent ;"""
table_data = pd.read_sql(query, engine)
display(table_data)
```

### Q10. Display all deals handled by sales agents in the East region (join sales_pipeline and sales_teams).

```python
query = f"""select sp.product,sp.account,sp.sales_agent,st.regional_office,sp.deal_stage from sales_teams as st 
            join sales_pipeline as sp 
            on st.sales_agent = sp.sales_agent
            where regional_office = 'East';"""
table_data = pd.read_sql(query, engine)
display(table_data)
```
