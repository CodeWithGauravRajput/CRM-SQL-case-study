```python
from sqlalchemy import create_engine
import pandas as pd
```

```python
engine = create_engine('mysql+pymysql://root:password@127.0.0.1/crm_sales')
```

```python
tables = ['accounts','data_dictionary','products', 'sales_pipeline', 'sales_teams']
```

```python
for table in tables:
    query = f"""select * from {table} limit 10;"""
    table_data = pd.read_sql(query, engine)
    print(f'{table} Table')
    display(table_data)
   
```

### Q1. List all companies (accounts) with their sector, revenue, and number of employees.

```python
query = f"""select account,sector,revenue,employees from accounts;"""
table_data = pd.read_sql(query, engine)
display(table_data)
```

### Q2. Display all unique sectors from the accounts table.

```python
query = f"""select DISTINCT sector from accounts;"""
table_data = pd.read_sql(query, engine)
display(table_data)
```

### Q3. Show the total number of accounts present in the CRM database.

```python
query = """select count(*) from accounts;"""
table_data = pd.read_sql(query,engine)
display(table_data)
```

### Q4. Find the average revenue of all companies.

```python
query = """select avg(revenue) as Avg_revenue from accounts;"""
table_data = pd.read_sql(query,engine)
display(table_data)
```

### Q5. Show all companies located in the United States.

```python
query = """select account,office_location from accounts where office_location = 'United States';"""
table_data = pd.read_sql(query,engine)
display(table_data)
```

### Q6. List all companies established before the year 2000.

```python
query = """select account,year_established from accounts where year_established < 2000;"""
table_data = pd.read_sql(query,engine)
display(table_data)
```

### Q7. Find the oldest company and its year of establishment.

```python
query = """select account,year_established from accounts order by  year_established Asc limit 1;"""
table_data = pd.read_sql(query,engine)
display(table_data)
```

### Q8. List all companies that are subsidiaries of another company.

```python
query = """select * from accounts WHERE subsidiary_of IS NOT NULL AND subsidiary_of <> '';"""
table_data = pd.read_sql(query, engine)
display(table_data)
```

###### if the query WHERE subsidiary_of IS NOT NULL returns rows, but those rows appear to have nothing in the subsidiary_of column, it's almost certainly because the column contains blank spaces, empty strings, or other invisible characters instead of a true NULL value.

### Q9. Show all medical sector companies and their revenue.

```python
query = """select account,sector,revenue from accounts where sector = 'medical';"""
table_data = pd.read_sql(query, engine)
display(table_data)
```

### Q10. Find the total number of employees across all companies.

```python
query = """select sum(employees) as Total_employees from accounts ;"""
table_data = pd.read_sql(query, engine)
display(table_data)
    
```
