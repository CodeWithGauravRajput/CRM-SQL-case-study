```python
from sqlalchemy import create_engine
import pandas as pd
```

```python
engine = create_engine('mysql+pymysql://root:password@127.0.0.1/crm_sales')
```

### 1. Find the top 3 sales agents by total close_value from Won deals.

```python
query = f"""select sales_agent,deal_stage,sum(close_value) as Total_vales from sales_pipeline WHERE deal_stage = 'Won' group by sales_agent order by Total_vales desc limit 3;"""
table_data = pd.read_sql(query, engine)
display(table_data)
```

### 2. Calculate each agent‚Äôs success rate = (Number of Won deals / Total deals) * 100.

```python
query = f"""SELECT 
    sales_agent,
    sum(deal_stage = 'Won') AS won_total_deals,
    COUNT(*) AS total_deals,
    (sum(deal_stage = 'Won') /COUNT(*))*100 as success_rate
    
FROM 
    sales_pipeline
GROUP BY 
    sales_agent;"""
table_data = pd.read_sql(query,engine)
display(table_data)
```

### 3. Find the average deal size for each product.

```python
query = f"""SELECT product,avg(close_value) as Avg_deal_size from sales_pipeline group by product;"""
table_data = pd.read_sql(query,engine)
display(table_data)
```

### 4. Identify which sector contributes the most total revenue .

```python
query = f"""SELECT sector,sum(revenue) as Total_revenue from accounts group by sector order by  Total_revenue desc limit 1;"""
table_data = pd.read_sql(query,engine)
display(table_data)
```

### 5. Determine which company has won the most deals.

```python
query = f"""SELECT account,count(deal_stage) as most_won_deals from sales_pipeline where deal_stage = 'Won' group by account order by most_won_deals desc limit 1 ;"""
table_data = pd.read_sql(query,engine)
display(table_data)
```

### 6. Calculate the average sales cycle duration in days (difference between engage_date and close_date).

```python
query = f"""SELECT (engage_date),close_date,close_date-engage_date from sales_pipeline;"""
table_data = pd.read_sql(query,engine)
display(table_data)
```

### 7. Find the month in which most deals were closed.

```python
query = f"""SELECT MONTHNAME(close_date) as Months,count(MONTHNAME(close_date)) as Month_counts from sales_pipeline group by MONTHNAME(close_date) order by Month_counts desc limit 1;"""
table_data = pd.read_sql(query,engine)
display(table_data)
```

### 8. Display the top-performing region (Central, East, West) based on Won deals‚Äô close_value.

```python
query = f"""SELECT st.regional_office,sum(sp.close_value) as Total_Won_close_value from sales_teams as st join sales_pipeline as sp on st.sales_agent = sp.sales_agent where sp.deal_stage = "Won"   group by st.regional_office order by Total_Won_close_value desc limit 1 ;"""
table_data = pd.read_sql(query,engine)
display(table_data)
```

### 9. Show total revenue per manager .

```python
query = f"""select st.manager,sum(ac.revenue) as Total_revenue from accounts as ac join sales_pipeline as sp on ac.account = sp.account join sales_teams as st  on sp.sales_agent =st.sales_agent group by st.manager ;"""
table_data = pd.read_sql(query,engine)
display(table_data)
```

### 10. List all deals where the product series is ‚ÄúGTX‚Äù and status is ‚ÄúWon‚Äù.

```python
query = f"""SELECT * from products as p join sales_pipeline as sp on p.product = sp.product where p.series = "GTX" and sp.deal_stage = "Won" ;"""
table_data = pd.read_sql(query,engine)
display(table_data)
```
